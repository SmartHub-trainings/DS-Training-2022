```python
df_mod=df.drop([  'marital', 'default', 'housing', 'loan',
       'contact', 'month', 'day_of_week', 'duration',  'pdays','poutcome' ],axis=1)
```

```python
df_mod.head()
```

```python
df_mod["target"]=df_mod.y.map({"no":0,"yes":1})
df_mod.drop("y",axis=1,inplace=True)
```

```python
df_mod.head()
```

```python
def change_job(job):
    if job=="unknown":
        return np.nan
    return job
df_mod.job=df_mod.job.apply(change_job)
```

```python
(df_mod.job=="unknown").sum()
```

```python
def change_education(education):
    if education=="unknown":
        return np.nan
    return education
df_mod.education=df_mod.education.apply(change_education)
```

```python
(df_mod.education=="unknown").sum()
```

```python
df_mod.fillna(method="bfill",inplace=True)
```

```python
df_mod.isna().sum()
```

```python
df_mod.head(1)
```

```python
df_numeric=pd.get_dummies(df_mod,columns=["job","education","marital_modified","loan_mod"],drop_first=True)
```

```python
df_numeric.head()
```

```python
df_numeric.info()
```

```python
X,y=df_numeric.drop("target",axis=1),df_numeric.target
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y)
```

```python
tree_model1=DecisionTreeClassifier()
tree_model1.fit(X_train,y_train)
print(tree_model1.score(X_train,y_train))
print(tree_model1.score(X_test,y_test))
```
