```python
import pandas as pd
```

```python
cov19=pd.read_csv("../../../datasets/covid-19/country_wise_latest.csv")
```

```python
cov19.head()
```

```python
cov19.describe()
```

```python
min_=cov19.Deaths.min()
max_=cov19.Deaths.max()
```

```python
min_
```

```python
max_
```

```python
((cov19.Deaths-min_)/(max_-min_)).describe()
```

```python
(cov19.Deaths-min_)/(max_-min_)
```

```python
cov19["Deaths_scaled"]=(cov19.Deaths-min_)/(max_-min_)
```

```python
cov19.head()
```

```python
mean_=cov19.Recovered.mean()
std_=cov19.Recovered.std()
```

```python
cov19["Recovered_scaled"]=(cov19.Recovered-mean_)/std_
```

```python
cov19.head()
```

```python
cov19.describe()
```

```python
cov19.Recovered_scaled.plot()
```
